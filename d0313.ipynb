## '인덱스'는 '무엇인가 가르킨다'라는 의미
## '슬라이싱'은 '잘라낸다' 라는 의미

```python
#변수에 저장된 값을 수정할 때
과일 = ['복숭아', '딸기', '수박', '사과']

과일[1]='포도'
print(과일)

#삭제

#1. del (예약어)
#포도를 삭제하고 싶다
del 과일[1]
print(과일)

#2. list 함수를 사용해서 삭제 :remove
과일.remove('사과')
print(과일)

#3. list 함수를 사용해서 삭제 :pop
#pop: 리스트에 맨 마지막 요소만 제거
과일.pop()
print(과일)
```

- 제어문 (조건문: if문)


if 조건식 :
   조건식이 참일 때 실행할 문장 (비교연산자 OR 논리연산자)
   
else:
   조건식이 거짓일 때 실행할 문장

   
   

```python
a = 12
if a >= 10 :
    print('a가 10보다 크거나 같다')
else:
    print('a가 10보다 작다')
```

```python
포켓 = ['핸드폰', '볼펜']
if '현금' in 포켓:
    print('택시타고 간다.')
elif '카드' in 포켓:
    print('버스타고 간다.')
else:
    print('걸어 간다.')
```
