```python
import random
from collections import Counter #class는대문자

num_draw = 2500
numbers = 6
lotto_range = 45

all_num = []

for _ in range(num_draw): 
    draw = random.sample(range(1, lotto_range+1), numbers)
    all_num.extend(draw )

counter = Counter(all_num)

final_num = [num for num, count in counter.most_common(6)]
final_num.sort()

print(f'{num_draw}번 추출 후 개수가 가장 많은 6개 숫자: ')
print(final_num)
print('\각 숫자 누적 횟수 : ')
for num in final_num:
    print(f'{num} : {counter[num]}회')
```
