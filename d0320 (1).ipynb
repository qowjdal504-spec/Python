### 판다스 데이터 확인

- df.head(n) : 상위 n개(비웠을 때 5개)의 데이터만 보겠다.
- df.tail() : 하위 5개의 데이터만 보겠다.
- df.info() : 컬럼명, 결측치 여부, 데이터 타입
- df.describe() : 숫자형 데이터의 통계량 (평균, 표준편차, 사분위수)
  이상치 --> 정규화(가장 일반적인 정규화 방법으로, 각 데이터를 최소값을 0, 최대값을 1로 변환), 표준화(데이터를 평균이 0 표준 편차가 1이 되도록 변환하는 과정)
- df.shape() : 데이터의 행과 열 크기

# 데이터의 집계 및 병합

***
- 그룹화 : 특정 기준에 따라 데이터를 묶어 통계
- ex) df.groupby('성별')['점수'].mean()
- 성별에 따른 점수 평균

```python
df
```

```python
# Type1이 Fire이면서 speed가 100 이상인 포켓몬의 이름과 Total 점수만 출력

df.loc[(df['Type 1']=='Fire') & (df['Speed']>=100),['Name', 'Total']]
```

```python
df
```

```python
# 포켓몬의 물리 공격력과 특수 공격력 중 더 높은 값을 선택하여 Max_attack 이라는 칼럼을 만들어주세요

df['Max_attack'] = df[['Attack', 'Sp. Atk']].max(axis = 1)
```

```python
df
```

```python
#'Total' 점수가 높은 순서대로 내림차순 정렬하되, 
# 점수가 같다면 'Name'을 알파벳 순으로 오름차순 정렬하여, 상위 10개 행만 출력하세요

df.sort_values(by = ['Total','Name'], ascending = [False,True]).head(10)
```

# 기본 그룹화 : groupby() + 집계함수

sum / count / size

- df.groupby('Type 1').mean()
- df.groupby('Type 1').size() : type 1 별로 포켓몬이 몇마리씩 있는지 출력

# 여러 개의 집계를 한번에 하는 방법

- df.groupby('Type 1').agg(['mean','min','max'])

# 다중 인덱스 그룹화
ex) Type1뿐만아니라 전설의 여부까지 포함해서 그룹화

- df.groupby(['Type 1','Legendary'])['Defense'].mean

```python
# 가장 강한 타입 찾기 내림차순으로 정렬

df.groupby('Type 1')['Total'].mean().sort_values(ascending=False)
```

```python
# 세대 별 포켓몬 수
df.groupby('Generation').size()
```

```python
# 타입 별 체력의 총 합

df.groupby('Type 1')['HP'].sum()
```

```python
# 전설의 포켓몬의 타입 별 분포

df[df['Legendary'] == True].groupby('Type 1').size().sort_values(ascending=False)
```
